# Sequence Mutation - Strings, Lists, Tuples

```python
a = ""      # an empty string also can be created with 0 characters
b = "This is a string" # string...space is also considered as a character 
c = " (1)"    # We can give integer also as a string
m = len(b)   # to get the length of the string
print(m)
n = b[0]   # to print character at that index
print(n)
p = b.count("s")   # to count how many times the character is repeated in the variable
print(p)
r = b[5:7]      # to print characters between 2:7(excluding 7) [Ex:- (:9) from start to 6 (2:) from 2 to end]   *SLICE OPERATOR*
print(r)
s = b.index("i")    # to print the index of first occurrence of the given string/character
print(s)
h = b.upper()     # to print the string in upper case
print(h)
v = b.strip()    # to print with spaces
print(v)
w = b.split("i") # it splits the strings with the feature provided ()-> space (Similarly works JOIN)
print(w)
X = a+b+c      #catenation
print(X)
 # Strings are immutable: they cannot be changed the edits we do are just a copy of it
```

Each pixel of the image will represent a single color. The specific color depends on a formula that mixes various amounts of three basic colors: red, green, and blue. This technique for creating color is known as the RGB Color Model. The amount of each color, sometimes called the intensity of the color, allows us to have very fine control over the resulting color.


import image                      # optional in course

img = image.Image("luther.jpg")
win = image.ImageWin(img.getWidth(), img.getHeight())
img.draw(win)
img.setDelay(1,15)   # setDelay(0) turns off animation

for row in range(img.getHeight()):
    for col in range(img.getWidth()):
        p = img.getPixel(col, row)

        newred = 255 - p.getRed()
        newgreen = 255 - p.getGreen()
        newblue = 255 - p.getBlue()

        newpixel = image.Pixel(newred, newgreen, newblue)

        img.setPixel(col, row, newpixel)

img.draw(win)
win.exitonclick()

```python
import turtle
wn = turtle.Screen()
apple = turtle.Turtle()
apple.begin_fill() # Begin the fill process.
apple.down() # "Pen" down?
for i in range(4):  # For each edge of the shape
    apple.forward(40) # Move forward 40 units
    apple.left(90) # Turn ready for the next edge
apple.up() # Pen up
apple.end_fill() # End fill.
```

**deterministic**
A process that is repeatable and predictable.

**documentation**
A place where you can go to get detailed information about aspects of your programming language.

**module**
A file containing Python definitions and statements intended for use in other Python programs. The contents of a module are made available to the other program by using the import statement.

**namespace**
A naming system for making names unique, to avoid duplication and confusion. Within a namespace, no two names can be the same.

**pseudo-random number**
A number that is not genuinely random but is instead created algorithmically.

**random number**
A number that is generated in such a way as to exhibit statistical randomness.

**random number generator**
A function that will provide you with random numbers, usually between 0 and 1.

**standard library**
A collection of modules that are part of the normal installation of Python.

```python
import random

howmany = 10
for counter in range(howmany):
    arandom = random.random()
    print(arandom)
```

```python
# print 10 random numbers between 25 and 35.
import random
howmany = 10
for counter in range(howmany):
    arandom = random.randrange(25,35)
    print(arandom)
```
